```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
belgrano = "A872811.xls"
losjuries = "A872956.xls"
malbran = "A872961.xls"
montequemado = "A872954.xls"
quimili = "A872955.xls"
sachayoj = "A872812.xls"
```

```python
belgrano_df = pd.read_excel(belgrano, sheet_name="Datos diarios")
belgrano_dfcleaned = belgrano_df.dropna(axis=1, how="all")
belgrano_dfcleaneddef = belgrano_dfcleaned.dropna(subset=["Radiacion_Global"])
belgrano_dfcleaneddef["Nombre_Estacion"] = "Bandera - EEA E Santiago"
belgrano_dfcleaneddef["Latitud"] = -28.89
belgrano_dfcleaneddef["Longitud"] = -62.27
belgrano_dfcleaneddef["Altura"] = 98
belgrano_dfcleaneddef.info()
belgrano_numerico = belgrano_dfcleaneddef.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 8))
plt.title("Correlacin de variables Belgrano")
sns.heatmap(
    belgrano_numerico.corr(), annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1, linewidths=.5, linecolor="black",annot_kws={'size': 8}
)
plt.show()
```
