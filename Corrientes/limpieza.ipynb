```python
# s to represent a pandas Series object;
# df to represent a pandas DataFrame object;
## Also: t – tuple, l – list, b – Boolean, i – integer,
# – numpy array, st – string, d – dictionary, etc

import pandas as pd
import matplotlib.pyplot as plt
import xlrd

# Lista de Enlaces (datasets)
urls = [
    "https://raw.githubusercontent.com/LovisottoSantiago/IA-RadiacionSolar-UNAJ/main/Corrientes/A872914.xls",
    "https://raw.githubusercontent.com/LovisottoSantiago/IA-RadiacionSolar-UNAJ/main/Corrientes/A872915.xls",
    "https://raw.githubusercontent.com/LovisottoSantiago/IA-RadiacionSolar-UNAJ/main/Corrientes/A872916.xls",
    "https://raw.githubusercontent.com/LovisottoSantiago/IA-RadiacionSolar-UNAJ/main/Corrientes/A872950.xls",
    "https://raw.githubusercontent.com/LovisottoSantiago/IA-RadiacionSolar-UNAJ/main/Corrientes/A872951.xls",
    "https://raw.githubusercontent.com/LovisottoSantiago/IA-RadiacionSolar-UNAJ/main/Corrientes/A872952.xls"
]

dataframes = []

for url in urls:
    df = pd.read_excel(url)
    dataframes.append(df)


# Unificar todos los datasets en uno solo
df_unificado = pd.concat(dataframes, ignore_index=True)

# Analizo la longitud del dataset para ver que tenga sentido, cada dataset tiene aprox 2000-4000 filas.
# print(len(df_unificado))  

# Ver que la columna Fecha tenga sentido en todo el rango.
# print(df_unificado['Fecha'].min())
# print(df_unificado['Fecha'].max())


# Guardar en Excel
df_unificado.to_excel("dataset_unificado_sin_limpiar.xlsx", index=False)
print("Se hizo un backup del dataset unificado (sin limpiar).")


# 1. Eliminar columnas con más del 90% de valores nulos (faltantes)
porcentaje_nulos = df_unificado.isnull().mean()

columnas_a_eliminar = porcentaje_nulos[porcentaje_nulos > 0.9].index

print(f"\nColumnas a eliminar por tener más del 90% nulos: {len(columnas_a_eliminar)}")
print(columnas_a_eliminar)

# Columnas eliminadas por tener más del 90% de nulos:
# ['Temperatura_Intemperie_5cm_Minima','Temperatura_Intemperie_50cm_Minima', 'Temperatura_Suelo_5cm_Media', 'Temperatura_Inte_5cm', 'Temperatura_Intemperie_150cm_Minima', 'Humedad_Suelo']


# Eliminamos esas columnas
print(f"\nColumnas antes de eliminar: {len(df_unificado.columns)}")
df_unificado = df_unificado.drop(columns=columnas_a_eliminar)
print("- Se eliminaron columnas con más del 90% de faltantes")
print(f"Columnas después de eliminar: {len(df_unificado.columns)}")

# print(df_unificado.columns.tolist()) # Para ver el nombre de las columnas
filas_sin_radiacion = df_unificado[df_unificado['Radiacion_Global'].isnull()]
print(f"\n- Cantidad de filas sin datos de Radiación Global: {len(filas_sin_radiacion)}")
print(f"Tamaño del dataset antes de eliminar: {len(df_unificado)}")

# Eliminar las filas por su índice
df_unificado = df_unificado.drop(filas_sin_radiacion.index)
print(f"Tamaño del dataset después de eliminar: {len(df_unificado)}")


# Guardar en Excel
df_unificado.to_excel("dataset_unificado_limpio.xlsx", index=False)
print("\nSe hizo un backup del dataset unificado y limpio.")
```
